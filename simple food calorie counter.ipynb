```python
# Simple Food Calorie Counter with Indian dishes

calorie_dict = {
    "apple": 52,
    "banana": 89,
    "orange": 62,
    "bread_slice": 79,
    "rice_cup": 206,
    "egg": 78,
    "chicken_breast": 165,
    "milk_cup": 103,
    "cheese_slice": 113,
    "butter_tbsp": 102,
    "dal_cup": 198,
    "roti": 120,
    "paneer_100g": 265,
    "buttermilk_cup": 99,
    "samosa": 130,
    "idli": 39,
    "vada": 150,
    "biryani_cup": 290,
    "chai_cup": 50
}

def calorie_counter():
    total_calories = 0
    print("Food Calorie Counter")
    print("Enter food items you ate (type 'done' to finish):")
    
    while True:
        food = input("Enter food item: ").lower()
        if food == "done":
            break
        if food in calorie_dict:
            try:
                servings = float(input("Enter number of servings: "))
                total_calories += calorie_dict[food] * servings
                print(f"Added {calorie_dict[food] * servings} calories.")
            except ValueError:
                print("Please enter a valid number for servings.")
        else:
            print("Food item not found in database. Try again.")
    
    print(f"Total calories consumed: {total_calories}")

if __name__ == "__main__":
    calorie_counter()
```
